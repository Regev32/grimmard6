```python
import os
import gzip
import pickle 
```

```python
dir = "./2014_6Locus"
zips = [(file, os.path.join(dir, file)) for file in os.listdir(dir) if file.endswith(".gz")]
```

```python
for name, zip_path in zips:
    with gzip.open(zip_path, "rb") as f:
        data = f.read()
    with open("./freqs/" + name[:-3], "wb") as f:
        f.write(data)
```

```python
def convert_hap(hap):
    alleles = hap.split("~")
    alleles = sorted(alleles)
    return '~'.join(alleles)
```

```python
os.makedirs("./freqs_dicts", exist_ok=True)

ferqs_files = os.listdir("./freqs/")
all_freqs = {}


for file in ferqs_files:
    print(file)
    freqs = {}
    with open(f"./freqs/{file}", "r") as f:
        lines = f.readlines()
        for line in lines:
            hap, _, prob = line.split(",")
            prob = float(prob)
            hap = convert_hap(hap)
            freqs[hap] = prob
    
    all_freqs[file[:-6]] = freqs
    
    with open(f"./freqs_dicts/{file[:-6]}.pkl", "wb") as f:
        pickle.dump(freqs, f)
```

```python
with open(f"./freqs_dicts/all_freqs.pkl", "wb") as f:
    pickle.dump(all_freqs, f)
```
